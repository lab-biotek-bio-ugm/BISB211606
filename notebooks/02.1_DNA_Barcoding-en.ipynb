The amplification used a pair of universal COI primers:

* Reverse primer (LCO1490) = `5'-GGTCAACAAATCATAAAGATATTGG-3'`

The PCR product was then submitted for Sanger sequencing. Sequencing was performed using both primers separately, producing two sequence reads: one forward read and one reverse read, each provided as an ABI trace file.

## Reading Sanger Sequencing Results & Sequence File Formats
In this practical session, we will use BioPython to read and write several file formats commonly used to store sequencing data.

BioPython provides many modules for handling biological data. One of the most frequently used modules is SeqIO, which allows us to read, write, and convert between different sequence file formats.

First, load the `SeqIO` library from BioPython:

```python
from Bio import SeqIO
from pathlib import Path
# If BioPython is not yet installed, please install it first by running the following command in the terminal:
# conda install -c conda-forge biopython
```

### Setting Up Input and Output Folders

In this notebook, we will define the input and output directories for the analysis using the Python library [`pathlib`](https://docs.python.org/3/library/pathlib.html).

```python
from pathlib import Path

# Set up paths
data_dir = Path('../data/raw')

# If the output of data_dir.is_dir() is False,
# adjust the value of data_dir to the correct location
assert data_dir.is_dir(), f"Folder {data_dir} not found. Please adjust the value of data_dir to the correct location!"

# Set up output directory
output_dir = Path('../results/02.1')
output_dir.mkdir(parents=True, exist_ok=True)

# Example usage of pathlib
example_path = data_dir / "1st_BASE_3766078_S1_F.ab1"  # add child directory or file using "/"
print(example_path, example_path.is_file())  # check whether the given path refers to a file
```

### ABI Files

An ABI trace file is the output generated from Sanger sequencing using capillary electrophoresis. It contains the raw sequencing data, including the chromatogram peaks and the PHRED quality scores used for base calling.

```python
# Parse an ABI file with BioPython and display the sequence ID
for record in SeqIO.parse(data_dir / "1st_BASE_3766078_S1_F.ab1", "abi"):
    print(record.id)
```

```python
# Print the sequence obtained from the read
print("Sequence length = " + str(len(record.seq)) + " bp")  # display the sequence length
print(record.seq)
```

```python
# Print the PHRED quality score for each nucleotide
print(record.letter_annotations)
```

### FASTQ
FASTQ files are similar in structure to FASTA files. However, unlike FASTA, a FASTQ file does not only store the sequence information, but also includes the sequencing quality scores (PHRED).

The following script will generate a new file in FASTQ format from the trace file we have just read:

```python
SeqIO.write(record, output_dir / f"{record.id}.fastq", "fastq-sanger")
```

Please open the FASTQ file that was generated in your folder (you can find it in the left panel) and examine its structure.

### Quality Trimming & FASTA

```python
# Parse abi file using BioPython, quality trimming with Mott’s algorithm
for record_trim in SeqIO.parse(data_dir / "1st_BASE_3766078_S1_F.ab1", "abi-trim"):
    print(record_trim.id)
```

```python
# Print the trimmed sequence
print("Sequence length = " + str(len(record_trim.seq)) + " bp")  # display the sequence length
print(record_trim.seq)
```

#### Discussion

* Is the sequence length the same after trimming?
* Why do raw Sanger sequencing data need to undergo quality control and trimming before further analysis?
* Try to find out: who is the author of Mott’s algorithm?

```python
### FASTA
SeqIO.write(record_trim, output_dir / f"{record_trim.id}.fasta", "fasta")
```

### Consensus Sequence
Next, we will generate a consensus sequence from the forward and reverse reads using a tool from the EMBOSS suite called `merger`.

> If EMBOSS is not yet installed, you can install it by running the following command in the terminal:

```python
%%bash
output_dir="../results/02.1"
merger $output_dir/3766078_S1_F.fasta $output_dir/3766079_S1_R.fasta -sreverse2 -outseq $output_dir/merged_S1.fasta -outfile $output_dir/merged_S1.merger
```
