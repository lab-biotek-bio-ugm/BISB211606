#### Conserved regions
Not all **regions** of homologous proteins show the same level of variation. Some parts of a protein are more **conserved** than others.

Now, let us try aligning the frataxin sequences from humans and the mosquito Aedes:

```python
%%bash
data_dir="../data/raw"
matcher $data_dir/frataxin.human.gb $data_dir/frataxin.aedes.gb stdout
```

#### Discussion

* Why are some regions not aligned?
* Why might one region be more similar than another?
* Is there anything special about amino acids 92, 100, 101, 104, 108, 111, 112, 122, and 124 in the human frataxin protein? _**#hint :** open the `.gb` file using a text editor._

Now, let us try modifying the alignment algorithm parameters by changing the **gap opening penalty** to 4:

```python
%%bash
data_dir="../data/raw"
matcher $data_dir/frataxin.human.gb $data_dir/frataxin.aedes.gb stdout -gapopen 4
```

#### Discussion

* Is this alignment result better than the previous one?

As we explored in the earlier section, an alignment algorithm can produce more than one possible alignment. However, **by default**, only one result is usually shown.

We can display three additional alternative alignments from the alignment above by adding the following parameter:

```python
%%bash
data_dir="../data/raw"
matcher $data_dir/frataxin.human.gb $data_dir/frataxin.aedes.gb stdout -alternatives 3
```

### Global alignments
Now, let us try aligning the last two sequences again using `needle`, which performs global alignment:

## Application: Identifying Exons and Introns

Sequence alignment can also be used to identify the exon and intron structure of a gene. In this exercise, we will compare the genomic sequence and the cDNA sequence of the Arabidopsis gene AT3G4905.

EMBOSS also provides a program specifically designed for analyzing mRNA-derived sequences against genomic DNA, such as Expressed Sequence Tags (ESTs).

```python
%%bash
data_dir="../data/raw"
est2genome $data_dir/AT3G4905.1.cdna.fasta $data_dir/AT3G4905.1.genom.fasta stdout -auto
```

#### Discussion

* What characteristics of mRNA make it necessary to use a more specialized alignment algorithm?
* How many introns are present in AT3G4905?

We can also obtain a more detailed alignment output by adding the following parameter:

```python
%%bash
data_dir="../data/raw"
est2genome $data_dir/AT3G4905.1.cdna.fasta $data_dir/AT3G4905.1.genom.fasta stdout -align -auto
```

#### Discussion
