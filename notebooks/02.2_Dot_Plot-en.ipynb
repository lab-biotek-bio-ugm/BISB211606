## 1. Dot Plot

The dot plot method is used to visually compare two sequences. In its simplest form, two sequences are compared using a matrix, where one sequence is placed on the vertical axis and the other on the horizontal axis. Whenever a match occurs between positions in the two sequences, a dot is placed at the corresponding coordinates in the matrix.

![simple dotplot](../assets/images/DOT_PLOT.png)

More advanced implementations of the dot plot method use **sliding windows** and **threshold values** to determine whether segments of the two sequences share sufficient similarity.

```python
# Load required libraries
from IPython.display import Image
from pathlib import Path

### Setting up input and output folders

# Set up paths
data_dir = Path('../data/raw')

# If data_dir.is_dir() returns False,
# adjust the value of data_dir to the correct location
assert data_dir.is_dir(), f"Folder {data_dir} not found. Please adjust the value of data_dir to the correct location!"

# Set up output directory
output_dir = Path('../results/02.2')
output_dir.mkdir(parents=True, exist_ok=True)
```

### 1.1 Gene Structure Analysis: Exons and Introns

We can examine the exon–intron structure of a gene visually using EMBOSS `dotmatcher`.

In the script below, we will compare the structure of a genomic region from Arabidopsis (AT3G52905.1) with one of its corresponding cDNA sequences.

_**Note:** the `%%bash` at the beginning of the cell indicates that the command is executed in the BASH terminal, not in Python._

```python
%%bash
data_dir="../data/raw"
output_dir="../results/02.2"
dotmatcher $data_dir/AT3G4905.1.genom.fasta $data_dir/AT3G4905.1.cdna.fasta -graph png -goutfile $output_dir/01_AT3G4905
```

```python
Image(filename=str(output_dir / '01_AT3G4905.1.png'))
```

### 1.2 Comparing Closely Related Genes

We can also explore **homologous regions** between two proteins, one from **human** and one from **dog**. Let us compare the **coding sequence (CDS)** of the **CDKL4 gene** between human and dog:

```python
%%bash
data_dir="../data/raw"
output_dir="../results/02.2"
dotmatcher -windowsize 15 \
    -threshold 35 \
    -graph png \
    -goutfile $output_dir/03_CDKL4 \
    $data_dir/CDKL4.human.fasta \
    $data_dir/cyclin.dog.fasta
```

```python
Image(filename=str(output_dir / '03_CDKL4.1.png'))
```

#### Discussion
